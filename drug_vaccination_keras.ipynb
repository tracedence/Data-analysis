```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
## generating training datasets
train_sample = []
train_label = []
for i in range(2000):
  young_people = randint(13,60)
  train_sample.append(young_people)
  train_label.append(0)

  older_people = randint(65, 100)
  train_sample.append(older_people)
  train_label.append(1)


```

```python
train_sample
```

```python
train_label
```

```python
train_sample = np.array(train_sample)
train_sample.shape
```

```python
train_label = np.array(train_label)
train_label.shape
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaler_train_sample = scaler.fit_transform(train_sample.reshape(-1,1))
scaler_train_sample.shape
```

```python
scaler_train_sample
```

```python
model = Sequential()
# input to the model will an array of (16,1)
model.add(tf.keras.Input(shape=1,))
## relu = sigmoid function
model.add(tf.keras.layers.Dense(32, activation='relu'))
## softmax is just a modification of sigmoid function
model.add(tf.keras.layers.Dense(2, activation='softmax'))
## used when output is 0 or 1
model.compile(Adam(lr=0.001), loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
model.fit(scaler_train_sample, train_label, batch_size=10, epochs=10)
```

```python
## test data
test_sample = []
test_label = []
for i in range(200):
  young_people = randint(13,60)
  test_sample.append(young_people)
  test_label.append(0)

  older_people = randint(65, 100)
  test_sample.append(older_people)
  test_label.append(1)

test_sample = np.array(test_sample)
test_label = np.array(test_label)
scaled_test_label = scaler.transform(test_sample.reshape(-1,1))
```

```python
predict = model.predict(scaled_test_label)
```

```python
confusion_matrix(test_label, np.argmax(np.array(predict), axis=1))
```

```python
accuracy_score(test_label, np.argmax(np.array(predict), axis=1))
```
