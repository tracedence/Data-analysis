```python
import pandas as pd
import numpy as np
```

```python
## df.isnull().sum()--> return number of missing values
## df.dropna() --> doesn't reset index
## df.fillna(method="bfill") --> will fill null values with value after null, "ffill" will fill null values with value before null.
```

```python
df = pd.DataFrame({'number': [1,2,3,np.nan, np.nan, np.nan, 8]})
df
```

##Checking missing value

```python
df.isnull().sum()
```

### three null value in a column

## Droping null values

```python
df.dropna()
```

### null values drop

## fill null value with last value after null rows

```python
df.fillna(method="bfill")
```

## fill null values with first value before null

```python
df.fillna(method="ffill")
```

## filling null value with mean value

```python
df.fillna(df["number"].mean())
```

## interpolate: estimate unknown datapoints between two known points.
### application
* mostly used for imputation missing values, when avg value doesn't fit.
* image processing, when expanding an image, we can estimate the pixel value with the help of neighbouring pixels.
* Impute missing value in time series data because in time series we like to fill missing value with previous one or two values. Example: temperature.







```python
df.interpolate(method='linear')
```
